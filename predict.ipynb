```python
!pip install numpy==1.22.2
!pip install pandas==1.4.4
!pip install NERDA==0.9.6
!pip install torch==1.13.1

import numpy as np
import pandas as pd 
from NERDA.models import NERDA
import torch
import os
from google.colab import drive
import nltk
nltk.download('punkt')

pd.set_option('display.max_rows', 500)
```

```python
model_v1 = torch.load('/content/Survice-BERT_V1_32-32.pt')
model_v3 = torch.load('/content/Survice-BERT_V3_32-32.pt')
```

```python
def merged_results(predict_result_v1_s, predict_result_v3_s):
  predict_result_merge = []

  for i in range(len(predict_result_v1_s[1][0])):
    if predict_result_v1_s[1][0][i] != predict_result_v3_s[1][0][i]:
      if predict_result_v1_s[1][0][i] in ['B-CASE', 'B-DISEASE', 'I-DISEASE']:
        predict_result_merge.append(predict_result_v1_s[1][0][i])
      else:
        predict_result_merge.append(predict_result_v3_s[1][0][i])
    else:
      predict_result_merge.append(predict_result_v3_s[1][0][i])

  return predict_result_merge
```

```python
sentence1 = 'However, since 4 January 2023 and as of 13 February 2023, 11 cases (10 confirmed and one probable) including eight deaths (Case Fatality Rate (CFR) 73%) have been reported across two divisions in Bangladesh.'
predict_result_v1 = model_v1.predict_text(sentence1)
predict_result_v3 = model_v3.predict_text(sentence1)

pd.DataFrame({'value':predict_result_v1[0][0], 'ner_v1':predict_result_v1[1][0] , 'ner_v3':predict_result_v3[1][0], 'ner_vm': merged_results(predict_result_v1, predict_result_v3)})
```

```python
sentence2 = 'From 1 November 2022 to 27 January 2023, a total of 559 cases of meningitis (of which 111 are laboratory confirmed), including 18 deaths (overall CFR 3.2%), have been reported from Zinder Region, southeast of Niger, compared to the 231 cases reported during 1 November 2021 to 31 January 2022.'
predict_result_v1 = model_v1.predict_text(sentence2)
predict_result_v3 = model_v3.predict_text(sentence2)

pd.DataFrame({'value':predict_result_v1[0][0], 'ner_v1':predict_result_v1[1][0] , 'ner_v3':predict_result_v3[1][0], 'ner_vm': merged_results(predict_result_v1, predict_result_v3)})
```
